## 1 Уравнений параболического типа
 
### Постановка задачи для уравнений параболического типа

Рассмотрим одномерное уравнение теплопроводности:

$$
\frac{\partial u}{\partial t}=a^{2} \frac{\partial^{2} u}{\partial x^{2}}, \quad 0<x<l, \quad t>0 .
$$

Если на границах $x=0$ и $x=t$ заданы значения искомой функции $u(x, t)$ в виде

$$
\begin{array}{ll}
u(0, t)=\phi_{0}(t), & t>0, \\
u(l, t)=\phi_{l}(t), & t>0,
\end{array}
$$

т.е. граничные условия первого рода, и, кроме того, заданы начальные условия при $t=0$

$$
u(x, 0)=\psi(x), \quad 0 \leq x \leq l,
$$

то задачу (1)-(4) называют первой начально-краевой задачей для уравнения теплопроводности (1).

Если на границах $x=0$ и $x=l$ заданы значения производных искомой функции по пространственной переменной

$$
\begin{aligned}
\frac{\partial u(0, t)}{\partial x} &=\phi_{0}(t), \quad t>0 \\
\frac{\partial u(l, t)}{\partial x} &=\phi_{l}(t), \quad t>0
\end{aligned}
$$

то есть граничные условия второго рода, то задачу $(1),(5),(6),(4)$ называют второй начально-краевой задачей для уравнения тепелопроводности.

Если на границах заданы линейные комбинации искомой функции и ее производной по пространственной переменной

$$
\begin{array}{cl}
\alpha \frac{\partial u(0, t)}{\partial x}+\beta u(0, t)=\phi_{0}(t), & t>0 \\
\gamma \frac{\partial u(l, t)}{\partial x}+\delta u(l, t)=\phi_{l}(t),= & t>0
\end{array}
$$

то есть граничные условия третьего рода, то задачу $(1),(7),(8),(4)$ называют третьей начально-краевой задачей для уравнения теплопроводности (1).

#### Метод конечных разностей

Нанесем на пространственно-временную область $0 \leq x \leq l, 0 \leq t \leq T$ конечно-разностную сетку $\omega_{h \tau}$

$$
\omega_{h \tau}=\left\{x_{j}=j h, j=0,1, \ldots, N ; \quad t^{k}=k \tau, k=0,1, \ldots, K\right\}
$$

с пространственным шагом $h=l / N$ и шагом по времени $\tau=T / K .$ Введем два временных слоя: нижний $t^{k}=k \tau$, на котором распределение искомой функции $u\left(x_{j}, t^{k}\right), j=0,1, \ldots, N$, известно (при $k=0$ распределение определяется начльным условием $(4)$ вида $\left.u\left(x_{j}, t^{0}\right)=\psi\left(x_{j}\right)\right)$ и верхний временной слой $t^{k+1}=(k+1) \tau$, на котором распределение искомой функции $u\left(x_{j}, t^{k+1}\right), j=0,1, \ldots, N$, подлежит определению.

Сеточной функцией задачи (1)- (4), имеющей обозначение $u_{j}^{k}$, назовем однозначное отображение целых аргументов $j, k$ в значения функции $u_{j}^{k}=u\left(x_{j}, t^{k}\right)$ Введем сеточные функции $u_{j}^{k}, u_{j}^{k=1}$, первая из которых известна, вторая - подлежит определению. Для ее определения в задаче (1)-(4) аппроксимируем дифференциальные операторы отношением конечных разностей:

$$
\begin{gathered}
\left.\frac{\partial u}{\partial t}\right|_{j} ^{k}=\frac{u_{j}^{k+1}-u_{j}^{k}}{\tau}+O(\tau) \\
\left.\frac{\partial^{2} u}{\partial x^{2}}\right|_{j} ^{k}=\frac{u_{j+1}^{k}-2 u_{j}^{k}+u_{j-1}^{k}}{h^{2}}+O\left(h^{2}\right) .
\end{gathered}
$$

Подставляя эти выражения в задачу (1)-(4), получим явную конечноразностную схему:

$$
\frac{u_{j}^{k+1}-u_{j}^{k}}{\tau}=a^{2} \frac{u_{j+1}^{k}-2 u_{j}^{k}+u_{j-1}^{k}}{h^{2}}+O\left(\tau+h^{2}\right),
$$

где

$$
k=0,1, \ldots, K-1, \quad j=1, \ldots, N-1,
$$

Начальные и граничные условия будут иметь вид:

$$
u_{0}^{k}=\phi_{0}\left(t^{k}\right), \quad u_{N}^{k}=\phi_{l}\left(t^{k}\right), \quad u_{j}^{0}=\psi\left(x_{j}\right),
$$

гДе

$$
k=0,1, \ldots, K, \quad j=0,1, \ldots, N,
$$

где для каждого $j$-го уравнения все значения сеточной функции известны, за исключением одного - $u_{j}^{k+1}$, которое может быть определено явно.

$$\frac{\partial u}{\partial t}=a \frac{\partial^{2} u}{\partial x^{2}}+b \frac{\partial u}{\partial x}, \quad a>0, \quad b>0$$


$$u_{x}(0, t)-u(0, t)=\exp (-a t)(\sin (b t)-\cos (b t)),$$

```python
#%matplotlib notebook
import numpy as np
import math
from math import sin, cos, exp, sqrt, cosh, log, pi
from matplotlib import pyplot as plt
from matplotlib.colors import LightSource
from matplotlib import cm
from IPython.display import display
import ipywidgets as ip_w
from scipy.integrate import odeint
from scipy.linalg import solve
```

```python
def u_x_0(x):
    return cos(x)

def main(u_x_0, a, b, t0, t1, tau, x0, x1, h):
    
    Jn = int((x1-x0)/h)
    Kn = int((t1-t0)/tau)
    
    J = np.linspace(x0, x1, Jn) #столбцы
    K = np.linspace(t0, t1, Kn) #строки
    
    a_u_x = np.zeros((len(K),len(J)))
    
    for i in range(len(J)):
        a_u_x[0,i] = u_x_0(J[i])
        
    for i in range(1,len(K)):
        for j in range(1,len(J)-1):
            a_u_x[i,j]=tau*(a_u_x[i-1,j+1]*((a/(h**2))+(b/(h)))+a_u_x[i-1,j]*(1/tau-2*a/(h**2)-b/(h))+a_u_x[i-1,j-1]*a/(h**2))
        a_u_x[i,0] = (a_u_x[i,1]-h*math.exp(-a*K[i])*(math.sin(b*K[i])-math.cos(b*K[i])))/(1+h)
        a_u_x[i,len(J)-1] = (a_u_x[i,len(J)-2]+h*math.exp(-a*K[i])*(math.sin(b*K[i])+math.cos(b*K[i])))/(1-h)
        
    return J, K, a_u_x
            
    
```

```python
X, T, W = main(u_x_0, 1, 1, 0, 3, 0.0001, 0, math.pi, 0.1)
```

```python
def plot_XTW(X,T,W,j):
    fig = plt.figure(figsize=(16,8))
    
    plt.subplot(121)    
    for i in range (len(T)):
        plt.plot(X,W[i,:],'o:',label='t[%s]=%s'%(i,np.round(T[i],4)))
    plt.xlabel('x')
    plt.ylabel('w')
    #plt.legend(loc='bottom', bbox_to_anchor=(0.5, -0.1))
    plt.legend(bbox_to_anchor=(-.4, 1), loc=2, borderaxespad=0.)
    
    plt.subplot(122)  
    plt.plot(X,W[j,:],'o:',label='t[%s]=%s'%(j,np.round(T[j],4)))
    plt.xlabel('x')
    plt.ylabel('w')    
```

## 4.2 Уравнения гиперболического типа

### Постановка задачи

Рассотрим волновое уравнение в области $0<x<l, \;\;t>0$ :

$$
\frac{\partial^{2} u}{\partial t^{2}}=a^{2} \frac{\partial^{2} u}{\partial x^{2}},
$$

Данное уравнение описывает, в частности, процесс малых поперечных колебаний струны. В этом случае $u(x, t)$ - поперечные перемещения струны, $a$ - скорость распространения малых возмущений в материале, из которого изготовлена струна.

Если концы струны движутся по заданным законам, то есть на концах заданы перемещения или значения искомой функциии, то первая начально-краевая задачача для волнового уравнения имеет вид

$$
\left\{\begin{array}{l}
\dfrac{\partial^{2} u}{\partial t^{2}}=a^{2} \dfrac{\partial^{2} u}{\partial x^{2}}, \quad 0<x<l, \quad t>0 \\
u(0, t)=\phi_{0}(t), \\
u(l, t)=\phi_{l}(t), \\
u(x, 0)=\psi_{1}(x), \quad 0 \leq x \leq l \\
\dfrac{\partial u(x, 0)}{\partial t}=\psi_{2}(x), \quad 0 \leq x \leq l
\end{array}\right.
$$

причем если концы струны жестко закреплены, то $\phi_{0}(t)=\phi_{l}(t)=0 .$

#### Конечно-разностная аппроксимация

Заменяя производные соответствуюІиим разностными аналогами получим явную схему:

$$
\begin{gathered}
\dfrac{u_{j}^{k+1}-2 u_{j}^{k}+u_{j}^{k-1}}{\tau^{2}}=a^{2} \dfrac{u_{j+1}^{k}-2 u_{j}^{k}+u_{j-1}^{k}}{h^{2}}+O\left(\tau^{2}+h^{2}\right) \\
j=1, \ldots, N-1 ; \quad k=1,2, \ldots
\end{gathered}
$$

Если в правой части использовать выражение для следующего временного слоя, то получим неявную схему:

$$
\begin{gathered}
\dfrac{u_{j}^{k+1}-2 u_{j}^{k}+u_{j}^{k-1}}{\tau^{2}}=a^{2} \dfrac{u_{j+1}^{k+1}-2 u_{j}^{k+1}+u_{j-1}^{k+1}}{h^{2}}+O\left(\tau^{2}+h^{2}\right) \\
j=1, \ldots, N-1 ; \quad k=1,2, \ldots
\end{gathered}
$$


#### Задание

Решить начально-краевую задачу для дифференциального уравнения гиперболического типа.

2. $\quad\dfrac{\partial^{2} u}{\partial t^{2}}+3 \dfrac{\partial u}{\partial t}=\dfrac{\partial^{2} u}{\partial x^{2}}+\dfrac{\partial u}{\partial x}-u+\sin x \exp (-t)$</br></br> $u(0, t)=\exp (-t), $</br></br> $u(x, 0)=\cos x $ </br></br>$ u(\pi, t)=-\exp (-t) $ </br></br>$ u_{t}(x, 0)=-\cos x .$

```python
 def u_x1_t(t):
    return -exp(-t)

def u_0_t(t):
    return exp(-t)

def ut_x_0(x, tau):
    return(1-tau)*cos(x)

def main_1(u_x_0, u_x1_t, u_0_t, ut_x_0, t0, t1, tau, x0, x1, h):
    
    Jn = int((x1-x0)/h)
    Kn = int((t1-t0)/tau)
    
    J = np.linspace(x0, x1, Jn) #столбцы
    K = np.linspace(t0, t1, Kn) #строки
    
    a_u_x = np.zeros((len(K),len(J)))
    
    for j in range(len(J)):
        a_u_x[0,j] = u_x_0(J[j])
        a_u_x[1,j] = ut_x_0(J[j],tau)
        
    for i in range(len(K)):
        a_u_x[i,0] = u_0_t(K[i])
        a_u_x[i,len(J)-1] = u_x1_t(K[i])
        
    for k in range(2,len(K)):
        for m in range(1,len(J)-1):
            a_u_x[k,m] = ((tau**2)*a_u_x[k-1,m+1]*(1+h)+a_u_x[k-1,m]*(2*(h**2)+3*tau*(h**2)-2*(tau**2)-(tau**2)*h-(tau**2)*(h**2))+a_u_x[k-1,m-1]*(tau**2)-a_u_x[k-2,m]*(h**2)+(tau**2)*(h**2)*math.exp(-K[k-1])*math.sin(J[m]))/((3*tau+1)*(h**2))
        
    return J, K, a_u_x
```

## 4.3 Уравнения эллиптического типа

### Постановка задачи

Рассмотрим уравнение Пуассона:

$$
\frac{\partial^{2} u}{\partial x^{2}}+\frac{\partial^{2} u}{\partial y^{2}}=f(x, y) \text {. }
$$

Если на границе Г расчетной области задана искомая функция, то говорят, что задана задача Дирихле:

$$
\left\{\frac{\partial^{2} u}{\partial x^{2}}+\frac{\partial^{2} u}{\partial y^{2}}=f(x, y), \quad(x, y) \in \Omega,\left.u(x, y)\right|_{\Gamma}=\phi(x, y), \quad(x, y) \in \Gamma .\right.
$$

Если на границе $\Gamma$ задается нормальная производная искомой функции, то соответствующая вторая краевая задача называется задачей Неймана для уравнения Пуассона:

$$
\left\{\begin{array}{l}
\frac{\partial^{2} u}{\partial x^{2}}+\frac{\partial^{2} u}{\partial y^{2}}=f(x, y), \quad(x, y) \in \Omega, \\
\left.\frac{\partial u(x, y)}{\partial n}\right|_{\Gamma}=\phi(x, y), \quad(x, y) \in \Gamma,
\end{array}\right.
$$

где $n$ - направление внешней нормали к границе $\Gamma$.

#### Конечно-разностная аппроксимация

Рассмотрим краевую задачу для уравениня Пуассона в прямоугольнике $x \in\left[0, l_{1}\right], y \in\left[0, l_{2}\right]$, на который наложим сетку

$$
\omega_{h_{1}, h_{2}}=\left\{x_{i}=i h_{1}, i=0, \ldots, N_{1} ; \quad y_{j}=j h_{2}, j=0, \ldots, N_{2}\right\} .
$$

Введем сеточную функцию $u_{i j}, i=0, \ldots, N_{1}, j=0, \ldots, N_{2}$ и аппроксимируем дифференциальную задачу во внутренних узлах с помощью отношения конечных разностей по следующей схеме:

$$
\begin{gathered}
\frac{u_{i+1, j}-2 u_{i, j}+u_{i-1, j}}{h_{1}^{2}}+\frac{u_{i, j+1}-2 u_{i, j}+u_{i, j-1}}{h_{2}^{2}}+O\left(h_{1}^{2}+h_{2}^{2}\right)=f\left(x_{i}, y_{j}\right), \\\\
i=1, \ldots, N_{1}-1, \quad j=1, \ldots, N_{2}-1 .
\end{gathered}
$$

Данная схема имеет второй порядок точности по переменным $x$ и $y$. 

#### Задание

Решить начально-краевую задачу для дифференциального уравнения эллиптического типа.

$$
\begin{aligned}
&\text { 2. } \dfrac{\partial^{2} u}{\partial x^{2}}+\dfrac{\partial^{2} u}{\partial y^{2}}=-2 \dfrac{\partial u}{\partial y}-3 u, \\\\
&\quad u(0, y)=\exp (-y) \cos y, \\\\

&\quad u\left(\frac{\pi}{2}, y\right)=0, \\\\
&\quad u\left(x, \frac{\pi}{2}\right)=0 
\end{aligned}
$$


```python
def u_x0(x):
    return cos(x)

def u_x1(x):
    return 0
    
def u_0y(y):
    return exp(-y)*cos(y)

def u_1y(y):
    return 0

def main_3(u_x0, u_x1, u_0y, u_1y, x0, x1, h1, y0, y1, h2):
    In = int((x1-x0)/h1)
    Jn = int((y1-y0)/h2)
    
    I = np.linspace(x0, x1, In) #столбцы
    J = np.linspace(y0, y1, Jn) #строки
        
    w=np.zeros((len(J),len(I)))
    X, Y = np.meshgrid(I, J)

    for i in range(len(I)):
            w[0,i] = u_x0(I[i])  #Lower Boundary
            w[len(J)-1,i] = u_x1(I[i]) #Upper Boundary

    for j in range(len(J)):
            w[j,0] = u_0y(J[j]) #left Boundary
            w[j,len(I)-1] = u_1y(J[j])  #Right Boundary 
                
            

    fig = plt.figure(figsize=(16,8))
    ax = fig.add_subplot(projection='3d')
    ax.plot(I, np.full(In, x1), w[Jn-1,:], zdir='z')
    ax.plot(I, np.full(In, x0), w[0,:], zdir='z')
    ax.plot(np.full(Jn, y1), J, w[:,In-1], zdir='z')
    ax.plot(np.full(Jn, y0), J, w[:,0], zdir='z')
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("W")
    
    N=In
    h=h1
    X, Y = np.meshgrid(I, J)
    
    
    N2=(N-1)*(N-1)
    A=np.zeros((N2,N2))
    ## Diagonal            
    for i in range (0,N-1):
        for j in range (0,N-1):           
            A[i+(N-1)*j,i+(N-1)*j]= (3*h**2-2*h-4)

    # LOWER DIAGONAL        
    for i in range (1,N-1):
        for j in range (0,N-1):           
            A[i+(N-1)*j,i+(N-1)*j-1]= 1   
    # UPPPER DIAGONAL        
    for i in range (0,N-2):
        for j in range (0,N-1):           
            A[i+(N-1)*j,i+(N-1)*j+1]= (2*h+1)   

    # LOWER IDENTITY MATRIX
    for i in range (0,N-1):
        for j in range (1,N-1):           
            A[i+(N-1)*j,i+(N-1)*(j-1)]= 1        


    # UPPER IDENTITY MATRIX
    for i in range (0,N-1):
        for j in range (0,N-2):           
            A[i+(N-1)*j,i+(N-1)*(j+1)]= 1
    Ainv=np.linalg.inv(A)   
    fig = plt.figure(figsize=(12,4));
    plt.subplot(121)
    plt.imshow(A,interpolation='none');
    clb=plt.colorbar();
    clb.set_label('Matrix elements values');
    plt.title('Matrix A ',fontsize=24)
    plt.subplot(122)
    plt.imshow(Ainv,interpolation='none');
    clb=plt.colorbar();
    clb.set_label('Matrix elements values');
    plt.title(r'Matrix $A^{-1}$ ',fontsize=24)

    fig.tight_layout()
    plt.show();
    
    r=np.zeros(N2)

    # vector r      
    for i in range (0,N-1):
        for j in range (0,N-1):           
            r[i+(N-1)*j]=0     
    # Boundary        
    b_bottom_top=np.zeros(N2)
    for i in range (0,N-1):
        b_bottom_top[i]=u_x0(I[i+1]) #Bottom Boundary
        b_bottom_top[i+(N-1)*(N-2)]=u_x1(I[i+1])# Top Boundary

    b_left_right=np.zeros(N2)
    for j in range (0,N-1):
        b_left_right[(N-1)*j]=u_0y(J[j+1]) # Left Boundary
        b_left_right[N-2+(N-1)*j]=u_1y(J[j+1])# Right Boundary

    b=b_left_right+b_bottom_top
    
    C=np.dot(Ainv,r-b)
    w[1:N,1:N]=C.reshape((N-1,N-1))

    fig = plt.figure(figsize=(16,8))
    ax = fig.add_subplot(111, projection='3d');
    
    
    ax.plot_wireframe(X, Y, w,color='r');
    
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("W")
    plt.show();
```

## 4 Метод переменных направлений

Рассмотрим следующую задачу:

$$
\left\{\begin{array}{l}
\dfrac{\partial u}{\partial t}=a\left(\dfrac{\partial^{2} u}{\partial x^{2}}+\dfrac{\partial^{2} u}{\partial y^{2}}\right)+f(x, y, t), \quad x \in\left(0, l_{1}\right), \quad y \in\left(0, l_{2}\right), \quad t>0, \\\\
u\left(x, 0, t\right)=\phi_{1}(x, t), \quad x \in\left[0, l_{1}\right], \\\\
u\left(x, l_{2}, t\right)=\phi_{2}(x, t), \quad x \in\left[0, l_{1}\right], \\\\
u\left(0, y, t\right)=\phi_{3}(y, t), \quad y \in\left[0, l_{2}\right], \\\\
u\left(l_{1}, y, t\right) =\phi_{4}(y, t), \quad y \in\left[0, l_{2}\right], \\\\
u(x, y, 0)=\psi(x, y), \quad x \in\left[0, l_{1}\right], \quad y \in\left[0, l_{2}\right] .
\end{array}\right.
$$

В схеме метода переменных направлений (МПН) шаг по времени разбивается на несколько шагов, число которых равно числу независи- мых пространственных переменных (в двумерном случае - на два). Ha каждом дробном временном слое один из пространственных дифференциальных операторов аппроксимируется неявно (по соотвтетствующему координатному направлению осуществляются скалярные прогонки), а в остальные - явно. на следующем дробном паге другой дифференциальных оператор аппроксимируется неявно, а остальные - явно и т.Д. ## 4.5 Метод дробных шагов

В отличие от метода переменных направлений, в методе дробнах шагов используются только неявные конечно-разностные операторы, что делает его абсолютно устойчивым в задачах, не содержащих смешанных производных. Для задачи (ссылка) схема метода дробных шагов имеет вид:

$$
\begin{aligned}
&\frac{u_{i j}^{k+1 / 2}-u_{i j}^{k}}{\tau}=\frac{a}{h_{1}^{2}}\left(u_{i+1 j}^{k+1 / 2}-2 u_{i j}^{k+1 / 2}+u_{i-1 j}^{k+1 / 2}\right)=\frac{f_{i j}^{k}}{2}, \\
&\frac{u_{i j}^{k+1}-u_{i j}^{k+1 / 2}}{\tau}=\frac{a}{h_{1}^{2}}\left(u_{i+1 j}^{k+1}-2 u_{i j}^{k+1}+u_{i-1 j}^{k+1}\right)=\frac{f_{i j}^{k+1}}{2},
\end{aligned}
$$

С помощью чисто неявной подсхемы осуществляются скалярные прогонки в направлении оси $x$ в количестве, равном $(J-1)$ в результате чего получается сеточная функция $u_{i j}^{k+1 / 2}$. На втором дробном шаге по времени осуществляются скалярные прогонки в направлении оси $y$ в количестве, равном $(I-1)$, в результате чего получается сеточная функция $u_{i j}^{k+1}$ Схема метода дробных шагов имеет порядок сходимости $O\left(\tau+h^{2}\right)$, то есть первый порядок по времени и второй по переменным $x$ и $y$.

## 6 Решение интегральных уравнений

Рассмотрим интегральное уравнение Фредгольма второго рода:

$$
u(x)-\int_{a}^{b} K(x, s) u(s) d s=f(x), \quad x \in[a, b],
$$

где $u(x)$ - искомая функция с областью определения $[a, b]$, а $K(x, s)-$ ядро уравнения.

Если функция $K(x, s)$ отвечает условию

$$
\int_{a}^{b} \int_{a}^{b} K^{2}(x, s) d x d s<\inf
$$

а однородное интегральное уравнение

$$
u(x)-\int_{a}^{b} K(x, s) u(s) d s=0
$$

имеет только нулевое решение, то исходное уравнение имеет при любой $f(x) \in L_{2}[a, b]$ только одно решение. Рассмотрим метод квадратур, который заключается в том, что интеграл аппроксимируется некоторой формулой численного интегрирования (например, формулами методов трапеций, Симпсона и т.д.). Для функции с непрерывной $m$-й производной введем в рассмотрение квадратурную формулу

$$
\int_{a}^{b} g(s) d s=\sum_{j=1}^{n} A_{j}^{(n)} g\left(s_{j}^{(n)}\right)+O\left(\frac{1}{n^{m}}\right)
$$

При $m=2$ это может быть составная формула трапеций или средних прямоугольников, при $m=4$ - составная формула Симпсона или Гаусса по двум узлам и т.д.

Будем считать, что решение рассматриваемого интегрального уравнения существует и единственно.

Аппроксимируя интеграл формулой (10), получим:

$$
u(x)-\sum_{j=1}^{n} A_{j}^{(n)} g\left(s_{j}^{(n)}\right)=f(x)+R_{n}(x), \quad x \in[a, b]
$$

Пренебрегая величиной $R_{n}(x)$, получим:

$$
u(x)-\sum_{j=1}^{n} A_{j}^{(n)} g\left(s_{j}^{(n)}\right)=f(x), \quad x \in[a, b]
$$

Примем функцию $u_{n}(x)$ за приближенное решение уравнения (9). Процесс решения уравнения (11) может быть сведен к решению системы линейных алгебраических уравнений

$$
U_{i}^{(n)}-\sum_{j=1}^{n} A_{j}^{(n)} K\left(s_{i}^{(n)}, s_{j}^{(n)}\right) U_{j}^{(n)}=f\left(s_{i}^{(n)}\right), \quad i=1, \ldots, n .
$$

Таким образом, можно выделить следующие этапы алгоритма численного решения уравнения (9):

- выбирается квадратурная формула

- строится и решается система линейных алгебраических уравнений

- по ее решению находится приближенное решение исходного интегрального уравнения в виде функции $u_{n}(x)$.
